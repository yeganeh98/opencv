#02 : OpenCV 

### import library

```python
import cv2
```

#####get the image with imread

```python
img = cv2.imread('images/rock.jpg')
```

######check the type!

```python
type(img)
```

##### get the image shape

```python
img.shape
```

```python
plt.imshow(img)
```

#####opencv is reading the channels as **BGR**
#####we will convert opencv to the channels of the photo

```python
img_fix = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img_fix)
```

#####scale it to gray and check the **shape**

```python
img_gray = cv2.imread('images/rock.jpg',cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(img_gray, cmap = 'gray')
```

###Resize the image

```python
img_new = cv2.resize(img_fix , (1000,400))
```

```python
plt.imshow(img_new)
```

###Resize with Ratio



```python
width_ratio = 0.5
height_ratio = 0.5
img_new2 = cv2.resize(img_fix,(0,0),img_fix , width_ratio,height_ratio)
```

```python
plt.imshow(img_new2)
```

```python
img_new2.shape
```

###FLIP on horizontal axis

```python
img3 = cv2.flip(img_fix , 0)
```

```python
plt.imshow(img3)
```

###FLIP on Vertical axis

```python
img4 = cv2.flip(img_fix , 1)
```

```python
plt.imshow(img4)
```

###FLIP on horizontal and Vertical

```python
img5 = cv2.flip(img_fix , -1)
```

```python
plt.imshow(img5)
```
